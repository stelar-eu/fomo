Goal: Plot the products and hazards dataset from AK to see if it is appropriate for our algorithms.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('/home/jens/ownCloud/Documents/3.Werk/0.TUe_Research/0.STELAR/1.Agroknow/data/products_and_hazards.csv')

df.Date = pd.to_datetime(df.Date)

df = df.set_index('Date')

df
```

```python
# There seems to be a great difference in means. Let's check that out

means = df.mean(axis=0)

plt.bar(np.arange(len(means)), means)
```

Yup... they clearly seem ordered based on total amount of incidents. With products first, then hazards.

```python
n = 100
tmp = means[:n]

plt.bar(np.arange(n), tmp)
```

```python
# Get and plot the global trend and quantiles

global_trend = df.mean(axis=1)
global_median = df.median(axis=1)

plt.plot(global_trend)
plt.plot(global_median)
```
